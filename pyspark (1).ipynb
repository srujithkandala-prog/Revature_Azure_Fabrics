```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("RDDAssignment") \
    .getOrCreate()

sc = spark.sparkContext
```

```python
emp_rdd = sc.parallelize([
    (101, "John", "IT", 50000),
    (102, "Alice", "HR", 60000),
    (103, "Bob", "IT", 70000),
    (104, "David", "Finance", 80000),
    (105, "Eva", "HR", 55000),
    (106, "Frank", "IT", 65000),
    (107, "Grace", "Finance", 90000),
    (108, "Helen", "HR", 62000)
])

emp_rdd.collect()
```

# 2. RDD Transformations
## 1. Use map() to extract employee names.

```python
rrd1=emp_rdd.map(lambda x: x[1])
rrd1.collect()
```

## 2. Use map() to create (emp_id, salary) pairs.

```python
rrd2=emp_rdd.map(lambda x: (x[0],x[3]))
rrd2.collect()
```

## 3. Use filter() to get employees with salary > 60000.

```python
rdd3=emp_rdd.filter(lambda x:(x[3]>60000))
rdd3.collect()
```

## 4. Use filter() to get only IT department employees.

```python
rdd4=emp_rdd.filter(lambda x: (x[2]=="IT"))
rdd4.collect()
```

## 5. Use distinct() to find unique departments.

```python
rdd5=emp_rdd.map(lambda x:x[2]).distinct()
rdd5.collect()
```

## 6. Create another RDD and perform union().

```python
new_rdd=sc.parallelize([
    (101,"John","IT",50000),
(102,"Alice","HR",60000),
(103,"Bob","IT",70000),
(104,"David","Finance",80000),
(105,"Eva","HR",55000),

])

rdd6=emp_rdd.union(new_rdd)
rdd6.collect()
```

## 7. Perform intersection() between two employee ID RDDs

```python
rrd7=emp_rdd.intersection(new_rdd)
rrd7.collect()
```

## 8. Use flatMap() to split department names into characters.

```python
rrd8=emp_rdd.flatMap(lambda x: x[2])
rrd8.collect()
```

## 9. Use mapValues() on Pair RDD to increase salary by 10%.

```python
rrd9=emp_rdd.map(lambda x: (x[0],x[3]))
rrd_sal=rrd9.mapValues(lambda y: (y+(y*0.10)))
rrd_sal.collect()
```

## 10. Use reduceByKey() to calculate total salary by department.

```python
rrd10=emp_rdd.map(lambda x: (x[2],x[3]))
rrd_red=rrd10.reduceByKey(lambda a,b : a+b)
rrd_red.collect()
```
