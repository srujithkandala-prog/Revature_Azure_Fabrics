# create dataframes

```python

from pyspark.sql import SparkSession
from pyspark.sql.functions import *
from pyspark.sql.types import *

spark = SparkSession.builder.appName("EmployeeDF").getOrCreate()

data = [
    (101,"John","IT",50000,"Chennai",30),
    (102,"Alice","HR",60000,"Bangalore",28),
    (103,"Bob","Finance",70000,"Mumbai",35),
    (104,"David","IT",80000,"Chennai",40),
    (105,"Eva","HR",55000,"Hyderabad",25),
    (106,"Frank","Finance",90000,"Mumbai",45)
]

columns = ["emp_id","emp_name","department","salary","city","age"]

emp_df = spark.createDataFrame(data, columns)
```

Section 1 – Basic DataFrame Operations

```python
# 1. Display complete DataFrame
emp_df.show()
```

```python
# 2. Display schema
emp_df.printSchema()
```

```python
# 3. Select Employee Name and Salary
emp_df.select("emp_name","salary").show()
```

```python
# 4. Display Employee ID, Name and Department
emp_df.select("emp_id","emp_name","department").show()
```

```python
# 5. Rename emp_name
emp_df.withColumnRenamed("emp_name","employee_name").show()
```

```python
# 6. Add bonus = 10% salary
emp_df.withColumn("bonus", col("salary")*0.10).show()
```

```python
# 7. Add dept_code
emp_df.withColumn(
    "dept_code",
    when(col("department")=="IT","D01")
    .when(col("department")=="HR","D02")
    .otherwise("D03")
).show()
```

```python
# 8. Remove age column
emp_df.drop("age").show()
```

```python
# 9. salary_after_hike = salary + 15%
emp_df.withColumn(
    "salary_after_hike",
    col("salary")*1.15
).show()
```

```python
# 10. Display column names
print(emp_df.columns)
```

Section 2 – Filtering Records

```python
# 1. Salary > 60000
emp_df.filter(col("salary") > 60000).show()
```

```python
# 2. IT employees
emp_df.filter(col("department")=="IT").show()
```

```python
# 3. Chennai employees
emp_df.filter(col("city")=="Chennai").show()
```

```python
# 4. Age > 30
emp_df.filter(col("age") > 30).show()
```

```python
# 5. Age between 25 and 35
emp_df.filter(col("age").between(25,35)).show()
```

```python
# 6. IT and salary > 60000
emp_df.filter(
    (col("department")=="IT") &
    (col("salary") > 60000)
).show()
```

```python
# 7. HR or Finance
emp_df.filter(
    col("department").isin("HR","Finance")
).show()
```

```python
# 8. Salary is NULL
emp_df.filter(col("salary").isNull()).show()
```

```python
# 9. Name starts with A
emp_df.filter(col("emp_name").startswith("A")).show()
```

```python
# 10. City contains 'i'
emp_df.filter(col("city").contains("i")).show()
```

Section 6 – GroupBy Operations

```python
# 1. Employee count by department

emp_df.groupBy("department").count().show()
```

```python
# 2. Average salary by department
emp_df.groupBy("department").avg("salary").show()
```

```python
# 3. Maximum salary by department
emp_df.groupBy("department").max("salary").show()
```

```python
# 4. Minimum salary by department
emp_df.groupBy("department").min("salary").show()
```

```python
# 5. Total salary by department
emp_df.groupBy("department").sum("salary").show()
```

```python
# 6. Department-wise count and average salary
emp_df.groupBy("department").agg(
    count("*").alias("emp_count"),
    avg("salary").alias("avg_salary")
).show()
```

```python
# 7. City-wise employee count
emp_df.groupBy("city").count().show()
# 8. City-wise average salary
emp_df.groupBy("city").avg("salary").show()
```

Section 7 – Null Handling

```python
# 1. Records with NULL salary
emp_df.filter(col("salary").isNull()).show()
```

```python
# . Replace NULL salary with 50000
emp_df.na.fill({"salary":50000}).show()
```

```python
# 3. Drop records with NULL values
emp_df.na.drop().show()
```

```python
# 4. Count NULL salary records
emp_df.filter(col("salary").isNull()).count()
```

```python
# . Replace NULL salary using department average
dept_avg = emp_df.groupBy("department") \
                 .agg(avg("salary").alias("avg_salary"))

df = emp_df.join(dept_avg, "department")

df = df.withColumn(
    "salary",
    when(col("salary").isNull(),
         col("avg_salary"))
    .otherwise(col("salary"))
)

df.drop("avg_salary").show()
```
